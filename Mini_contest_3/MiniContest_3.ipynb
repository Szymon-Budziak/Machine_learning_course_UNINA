## Scaling data

#### resnet34

```python
resnet34_scaler = MinMaxScaler()
```

```python
resnet34_scaler.fit(resnet34_X)
```

```python
scaled_features = resnet34_scaler.fit_transform(resnet34_X)
```

## Train test split

```python
from sklearn.model_selection import train_test_split
```

#### efficientNet

```python
efficientNet_X_train, efficientNet_X_test, efficientNet_y_train, efficientNet_y_test = train_test_split(
    efficientNet_X, efficientNet_y, test_size=0.25, random_state=42, stratify=efficientNet_y, shuffle=True
)
```

#### mobileNet

```python
mobileNet_X_train, mobileNet_X_test, mobileNet_y_train, mobileNet_y_test = train_test_split(
    mobileNet_X, mobileNet_y, test_size=0.25, random_state=42, stratify=mobileNet_y, shuffle=True
)
```

#### resnet34

```python
resnet34_X_train, resnet34_X_test, resnet34_y_train, resnet34_y_test = train_test_split(
    resnet34_X, resnet34_y, test_size=0.25, random_state=42, stratify=resnet34_y, shuffle=True
)
```

## MLPClassifier

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
```

```python
clf = MLPClassifier(max_iter=3000)
```

Best parameters after doing first GridSearchCV.

Now i will change some parameters to find new best ones like hidden_layer_sizes or aplha.

```python
parameters = {
    'hidden_layer_sizes': ((100,), (100, 100, 100), (50, 150, 150, 50), (200, 200, 200, 200)),
    'activation': ['relu'],
    'solver': ('lbfgs', 'adam'),
    'alpha': (1, 10, 50, 100),
    'learning_rate': ['invscaling'],
}
```

```python
grid_search = GridSearchCV(estimator=clf,
                           param_grid=parameters,
                           n_jobs=-1,
                           cv=5,
                           scoring='accuracy')
```

#### efficientNet

```python
grid_search.fit(efficientNet_X_train, efficientNet_y_train)
```

```python
print(grid_search.best_params_)
clf_efficientNet_best = grid_search.best_estimator_
```

```python
pred = clf_efficientNet_best.predict(efficientNet_X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(confusion_matrix(efficientNet_y_test, pred))
```

```python
print(classification_report(efficientNet_y_test, pred))
```

```python
print(accuracy_score(efficientNet_y_test, pred))
```

#### mobileNet

```python
grid_search.fit(mobileNet_X_train, mobileNet_y_train)
```

```python
print(grid_search.best_params_)
clf_mobileNet_best = grid_search.best_estimator_
```

```python
pred = grid_search.predict(mobileNet_X_test)
```

```python
print(confusion_matrix(mobileNet_y_test, pred))
```

```python
print(classification_report(mobileNet_y_test, pred))
```

```python
print(accuracy_score(mobileNet_y_test, pred))
```

#### resnet34
