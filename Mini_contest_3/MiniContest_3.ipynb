```python
scaled_features = efficientNet_scaler.fit_transform(efficientNet_X)
```

#### resnet34

## Train test split

```python
from sklearn.model_selection import train_test_split
```

#### efficientNet

```python
efficientNet_X_train, efficientNet_X_test, efficientNet_y_train, efficientNet_y_test = train_test_split(
    efficientNet_X, efficientNet_y, test_size=0.25, random_state=100, stratify=efficientNet_y, shuffle=True)
```

#### mobileNet

```python
mobileNet_X_train, mobileNet_X_test, mobileNet_y_train, mobileNet_y_test = train_test_split(
    mobileNet_X, mobileNet_y, test_size=0.25, random_state=100, stratify=mobileNet_y, shuffle=True)
```

#### resnet34

```python
resnet34_X_train, resnet34_X_test, resnet34_y_train, resnet34_y_test = train_test_split(
    resnet34_X, resnet34_y, test_size=0.25, random_state=100, stratify=resnet34_y, shuffle=True)
```

#### efficientNet

```python
clf_efficientNet_best = MLPClassifier(max_iter=5000, activation='relu', learning_rate='invscaling', solver='lbfgs', alpha=10, hidden_layer_sizes=(100, 100, 100), random_state=100)
```

```python
clf_efficientNet_best.fit(efficientNet_X_train, efficientNet_y_train)
```

```python
pred = clf_efficientNet_best.predict(efficientNet_X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(confusion_matrix(efficientNet_y_test, pred))
```

```python
print(classification_report(efficientNet_y_test, pred))
```

```python
print(accuracy_score(efficientNet_y_test, pred))
```

#### mobileNet

```python
clf_mobileNet_best = MLPClassifier(max_iter=5000, activation='relu', learning_rate='invscaling', solver='lbfgs', alpha=55, hidden_layer_sizes=(100,), random_state=100)
```
